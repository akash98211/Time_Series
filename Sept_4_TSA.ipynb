```python
Timeseries : It is a set of oservations/datapoints taken at equal interval where our main
    variable will be time dependent. 
```

```python
In TSA, we need 2 variables i.e. time, Target variable 
```

```python
Types of TSA:
    1. Univariate Analysis(1 target var, 1 time dependent var)
        Autoregressive(AR)
        Moving Average(MA)
        ARMA
        ARIMA(Integrated)
        SARIMA(Seasonal)
        LSTM(Long Short Term Memory)
        fbprophet
    2. Multivariate Analysis(multiple var)
        VAR(Vector)
        SARIMAX
        LSTM
        FBProphet
```

```python
when we plot Data:
    on x axis : time
    y         : magnitude/values/amount
```

```python
limitations:
    Time dependent var is required
    Constant interval of time is required
```

```python
interval:
    an hour, a day, a week, month, quarter, yearly, decade, century
```

```python
Importance:
    1. Business Understanding/forecasting
    2. Understand Past behaviour
    3. Plan future
    4. Evaluate Current Accomplishment
```

```python
Components of TS:
    1. Trend
    2. Seasonality
    3. Irregularity
    4. Cyclic
```

```python
When should we not use TSA:
    1. Values are constant
    2. Time interval is not constant
```

```python
Stationarity:
    mean and variance is constant over a period of time and does not vary much
    TS assumes that data is stationary
```

```python
How to check stationarity:
    1. ADCF test(Augmented Dickey-Fuller Test)(pvalue<0.05)== Data is stationary
    2. Rolling Statistics
```

```python
How to make data Stationary:
    By create a new feature:
        1. differencing(shift value and take diff)
        2. rolling mean(window)
        3. Transformations:
            log 
            cbrt
            sqrt
            rcpr
```
